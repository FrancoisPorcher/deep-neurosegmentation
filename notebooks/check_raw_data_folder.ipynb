```python
import glob
from config import config  # Make sure your project structure and PYTHONPATH are set up correctly for this import to work

def count_nii_gz_files(dbb_path):
    """
    Counts all files with a .nii.gz extension in the specified directory.

    Parameters:
    dbb_path (str): Path to the directory to search for .nii.gz files.

    Returns:
    int: The count of .nii.gz files found.
    """
    # Construct the search pattern to find all .nii.gz files within dbb_path
    search_pattern = f"{dbb_path}/**/*.nii.gz"
    
    # Use glob to find all files matching the pattern, including in subdirectories
    files_found = glob.glob(search_pattern, recursive=False)
    
    # Return the count of files found
    return len(files_found)

# Example usage
dbb_path = config.DBB_PATH
nii_gz_file_count = count_nii_gz_files(dbb_path)
print(f"Found {nii_gz_file_count} files with a .nii.gz extension in {dbb_path}.")
```
